### 21. Determine and comment on the day of the week that movies are released. What is notable about release days and months throughout the year? (Use pd.pivot_table)

```python
dates['weekday'] = pd.to_datetime(dates['date']).dt.day_name()
dates['month'] = pd.to_datetime(dates['date']).dt.month_name()

pivot = pd.pivot_table(dates, index='month', columns='weekday', values='title', aggfunc='count').fillna(0)
pivot
```

Friday and Thursday are the most common days for movie releases, in that order. In terms of months, March and October stand out likely due to spring break and Halloween season.

### 22. Can you plot the pivot table? Is the output "correct"? Can you fix it?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday', 'Total']
pivot = pivot[[col for col in weekday_order if col in pivot.columns]]

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'Total']
pivot = pivot.reindex([month for month in month_order if month in pivot.index])

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='coolwarm', linewidths=0.9)
plt.title('Movie Releases by Month and Weekday')
plt.xlabel('Weekday')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```
